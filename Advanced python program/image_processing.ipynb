```python
def myCvtColor(img):
    img_new=np.mean(img,axis=2)
    img_new=np.array(img_new,dtype=np.uint8)
    print(img.shape,img_new.shape)
    return img_new
```

Color Segmentation

```python
import cv2
import numpy as np

frame=cv2.imread(r"C:\Users\ASUS\Downloads\Samples-20260613T124925Z-3-001\Samples\flower.jpg")
hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

lower_color = np.array((20,188,130))
upper_color = np.array((40,255,255))

myMask = cv2.inRange(hsv, lower_color, upper_color)

res = cv2.bitwise_and(frame,frame,mask=myMask)

cv2.imshow('res',res)
cv2.imshow('mask',myMask)
cv2.imshow('hsv',hsv)
cv2.imshow('frame',frame)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

Finding the color range

```python
import cv2
import numpy as np

#cap = cv2.VideoCapture(1)

def nothing(x):
    pass
#Creating a window for later use
cv2.namedWindow('result')

#starting with 100's to prevent error while masking
h,s,v = 100,100,100

#Creating track bar
cv2.createTrackbar('h','result',0,179,nothing)
cv2.createTrackbar('s','result',0,255,nothing)
cv2.createTrackbar('v','result',0,255,nothing)

while(1):

    frame = cv2.imread(r"C:\Users\ASUS\Downloads\Samples-20260613T124925Z-3-001\Samples\flower.jpg")
    #ret,frame=cap.read()
    frame=cv2.resize(frame,(650,350))

    #converting to HSV
    hsv = cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)

    #get info from track bar and appy to result
    h = cv2.getTrackbarPos('h','result')
    s = cv2.getTrackbarPos('s','result')
    v = cv2.getTrackbarPos('v','result')

    #Normal masking algorithm
    lower_color = np.array((h,s,v))
    upper_color = np.array((255,255,255))

    mask = cv2.inRange(hsv, lower_color, upper_color)

    result = cv2.bitwise_and(frame,frame,mask=mask)

    cv2.imshow('result',result)

    k = cv2.waitKey(5) & 0xFF
    if k == 27:
        break

cv2.destroyAllWindows()

```

Countour Operations

```python
import cv2
import numpy as np

# read image and prepare threshold
img = cv2.imread(r"C:\Users\ASUS\Downloads\Samples-20260613T124925Z-3-001\Samples\flower.jpg")
imgray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(imgray, 80, 255,0)

# find contours
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

# draw contours that satisfy area constraint
for cnt in contours:
    area = cv2.contourArea(cnt)
    if area > 2000 and area < 6000:
        cv2.drawContours(img, [cnt], -1, (0,255,0), 2)

# show threshold and image with selected contours
cv2.imshow('thresh', thresh)
cv2.imshow('img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

openCV Drawing Functions

```python
import cv2
import numpy as np
import os

# Image path
image_path = r"C:\Users\ASUS\OneDrive\Desktop\python new\codes\Samples\gray_image.png"

# Check whether file exists
if not os.path.exists(image_path):
    print("Image file not found!")
    print("Check the path:", image_path)

else:
    # Read image
    img = cv2.imread(image_path)

    if img is None:
        print("Failed to load image!")
    else:
        # Convert to grayscale
        imgray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

        # Threshold
        ret, thresh = cv2.threshold(imgray, 80, 255, 0)

        # Find contours
        contours, hierarchy = cv2.findContours(
            thresh,
            cv2.RETR_TREE,
            cv2.CHAIN_APPROX_SIMPLE
        )

        print("Number of contours:", len(contours))

        # Draw contours and bounding boxes
        for cnt in contours:
            area = cv2.contourArea(cnt)

            if 2000 < area < 6000:
                cv2.drawContours(img, [cnt], -1, (0, 255, 0), 2)

                x, y, w, h = cv2.boundingRect(cnt)

                print("x =", x, "y =", y, "w =", w, "h =", h)

                cv2.rectangle(
                    img,
                    (x, y),
                    (x + w, y + h),
                    (0, 0, 255),
                    2
                )

        # Display images
        cv2.imshow("Threshold", thresh)
        cv2.imshow("Contours", img)

        cv2.waitKey(0)
        cv2.destroyAllWindows()
```
